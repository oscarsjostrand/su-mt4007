### REST API

In this seqton we will fetcvh data from the Nobel Price REST API to study the price motivations of winners. We are interested in the word frequences and we will visualize this in a word cloud. 

```python
import requests
import json

import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
url = 'https://api.nobelprize.org/2.1/nobelPrizes'
params = {"sort" : "desc", "nobelPrizeCategory" : "phy"}
r = requests.get(url, params)
data = r.json()

data_list = data["nobelPrizes"]
data_list[0]
```

We now have the data in the form of a list. Now, let us extract the motivations and create a wordcloud using the wordcloud library. 

```python
motivations = []
# extract the motivations from data_list
for i in data_list:
    if 'laureates' in i:
        for laureate in i['laureates']:
            motivation = laureate.get('motivation', {}).get('en')
            if motivation:
                motivations.append(motivation)

                
stopwords = set(STOPWORDS)

motivations_text = " ".join(motivations)

# Create a WordCloud object
wordcloud = WordCloud(
    width=800, 
    height=400, 
    background_color='white', 
    stopwords = stopwords,
).generate(motivations_text)

# Display the word cloud
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
